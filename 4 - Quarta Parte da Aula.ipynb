```python
# Bibliotecas que serão utilizadas
from selenium import webdriver
from openpyxl import load_workbook
from openpyxl import Workbook
import time
```

```python
# Criação da tabela
tabela_base = Workbook()
tabela_base_sheet = tabela_base.active

# Criação da primeira linha com o cabeçalho da tabela
primeira_linha = ['ID', 'Escola', 'Codigo INEP', 'Estado', 'Cidade', 'Tipo', 'Qtd Alunos']
tabela_base_sheet.append(primeira_linha)
```

```python
# Instância do browser juntamente com o carregamento da página a ser raspada
firefox = webdriver.Firefox(executable_path='/usr/local/Cellar/geckodriver/0.26.0/bin/geckodriver')
firefox.get('http://educacaoconectada.mec.gov.br/consulta-pdde')
```

```python
# Localização do elemento para seleção do Estado
element_uf = firefox.find_element_by_id('filter_uf')
# Laço for percorre até encontrar 'DF'
for option in element_uf.find_elements_by_tag_name('option'):
    if option.text == 'DF':
        option.click()
        break

# Localização do elemento para seleção do Munícipio
element_municipio = firefox.find_element_by_id('filter_municipio')
# Laço for percorre até encontrar 'Brasilia'
for option in element_municipio.find_elements_by_tag_name('option'):
    if option.text == 'BRASILIA':
        option.click()
        break

# Localização do botão da Consulta
button_consultar = firefox.find_element_by_id('filter')
button_consultar.click()

```

```python
# Localização do dropdown que exibe 100 "caixas" por tela
element_data_table_length = firefox.find_element_by_name('data-table_length')
for option in element_data_table_length.find_elements_by_tag_name('option'):
    if option.text == '100':
        option.click()
        break

# Aguarda 10 segundos até que carregue a página
time.sleep(10)

```

```python
# Localização da tabela com os elementos que queremos raspar
table_total = firefox.find_element_by_id('data-table')
# Cada escola = um td
table_td = table_total.find_elements_by_tag_name('td')

print("page 1")
counter = 0
counterTotal = 0
citiesList = []

# Percorre todas as escolas exibidas em uma página
for row in table_td:
    # Sabemos que cada escola tem 6 atributos, por isso o counter % 6 == 0. Quando termina de colher 6 elementos ele salva em uma linha e parte para o próximo.
    counter += 1
    citiesList.append(row.text)
    if counter % 6 == 0:
        counterTotal += 1
        citiesList.insert(0, counterTotal)
        tabela_base_sheet.append(citiesList)
        print(citiesList)
        citiesList = []

# Localização do botão que vai para a próxima página
next_button = firefox.find_element_by_link_text('2')
next_button.click()

# Aguarda 10 segundos até que carregue a página
time.sleep(10)
```
